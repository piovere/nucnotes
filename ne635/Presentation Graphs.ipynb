```python
%matplotlib inline
import matplotlib.pyplot as plt
```

# Standards Measurements

```python
true = {
    'Burma spinel': 0.0020499,
    'CuO2': 0.0020360,
    'UO2': 0.0019591
}
```

```python
itu_measured = {
    'Burma spinel': 0.001984,
    'CuO2': 0.001974,
    'UO2': 0.001886
}
itu_errors = {
    'Burma spinel': 0.000012,
    'CuO2': 0.000004,
    'UO2': 0.000006
}
```

```python
llnl_measured = {
    'Burma spinel': 0.001986,
    'UO2': 0.001888
}
llnl_errors = {
    'Burma spinel': 0.000017,
    'UO2': 0.000007
}
```

```python
plt.plot(true.keys(), true.values(), '^', label='Standards')
plt.errorbar(itu_measured.keys(), itu_measured.values(), yerr=itu_errors.values(), 
             marker='o', linestyle='None', label='ITU', alpha=0.5)
plt.errorbar(llnl_measured.keys(), llnl_measured.values(), yerr=llnl_errors.values(), 
            marker='s', linestyle='None', label='LLNL', alpha=0.5)
plt.legend(loc="upper right")
plt.title(r"$^{18}\mathrm{O}/^{16}\mathrm{O}$ with $2\sigma$ error values")
plt.savefig('known_sample_results.png', dpi=300)
plt.show()
```

# IMF

```python
fig, axs = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(15, 5))

ax = axs[0]
ax.errorbar(['Burma spinel', r'$\mathrm{UO_2}$', r'$\mathrm{CuO_2}$'], 
            [1.033, 1.039, 1.032], yerr=[0.007, 0.003, 0.002], 
             label='ITU', linestyle='None', marker='o', color='orange')
ax.set_title('ITU')

ax = axs[1]
ax.errorbar(['Burma spinel', r'$\mathrm{UO_2}$'], 
            [1.032, 1.038], yerr=[0.009, 0.004], 
             label='LLNL', linestyle='None', marker='s', color='green')
ax.set_title('LLNL')

fig.suptitle("Instrumental Mass Discrimination Factor")
#fig.legend(loc='lower center')

plt.savefig('imf.png', dpi=300)
plt.show()
```
