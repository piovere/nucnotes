```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

# Problem 1

```python
keff = np.loadtxt('problem1/keff.dat')
burnup = np.loadtxt('problem1/burnup.dat')
plt.plot(burnup, keff)
plt.xlabel(r'Burnup $\frac{MWd}{Day}$')
plt.ylabel(r'$k_{eff}$')
plt.show()
```

```python
plt.plot(burnup, (keff - 1.) / keff)
plt.plot(burnup, [0 for b in burnup], 'r--')
plt.xlabel(r'Burnup')
plt.ylabel(r'Default configuration')
plt.title("All Rods Out")
plt.show()
```
