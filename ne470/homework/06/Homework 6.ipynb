```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

# Problem 1

```python
keff = np.loadtxt('problem1/keff.dat')
burnup = np.loadtxt('problem1/burnup.dat')
plt.plot(burnup, keff)
plt.xlabel(r'Burnup $\frac{MWd}{MTM}$')
plt.ylabel(r'$k_{eff}$')
plt.savefig('images/k_eff_normal.png')
plt.show()
```

```python
plt.plot(burnup, (keff - 1.) / keff)
plt.plot(burnup, [0 for b in burnup], 'r--')
plt.xlabel(r'Burnup $\frac{MWd}{MTM}$')
plt.ylabel(r'Reactivity')
plt.title("Default configuration")
plt.savefig('images/excess_reactivity_normal.png')
plt.show()
```
