```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as la
```

# 6.18c

```python
x = np.array([0.6598, 1.1699, 1.2316, 1.7602, 2.3432])
y = np.array([1.3620, 2.3908, 3.5360, 5.0183, 5.4835])
a, b = np.polyfit(x, y, 1)
print("y = a*x + b = {} * x + {}".format(a, b))
```

```python
xs = np.linspace(0, 3, 1000)
plt.plot(x, y, 'bo', label="Data")
plt.plot(xs, a*xs + b, label="LS Fit")
plt.plot(xs, 2*xs + 1, label="True curve")
plt.legend(loc="upper left")
plt.title("Problem 6.18c")
plt.show()
```
