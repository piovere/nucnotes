```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def y(x):
    return (1 - x ** 2) ** 0.5
```

```python
dat = np.array([[x, y(x)] for x in np.random.rand(4000)])
dat
```

```python
ax = plt.figure(figsize=(6,5))

plt.subplot(111)
plt.plot(dat[:,0], dat[:,1], ".")
plt.show()
```

```python
np.mean(np.array([y(x) for x in np.random.rand(1000000)]))
```

```python
np.pi / 4
```

```python
l = 1000
dat2 = [np.mean(y(np.random.rand(i))) for i in range(1, l)]

plt.figure(figsize=(5,5))

plt.subplot(111)
plt.plot(dat2, ".")
plt.plot([np.pi / 4] * l, "r--")
plt.show()
```
