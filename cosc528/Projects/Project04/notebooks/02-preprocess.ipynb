```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
train_file = '../data/processed/train.csv'
```

```python
train = pd.read_csv(train_file)
```

```python
pred_cols = train.columns[:-1]
```

```python
scaler = StandardScaler().fit(train[pred_cols])
```

```python
train[pred_cols] = scaler.transform(train[pred_cols])
```
