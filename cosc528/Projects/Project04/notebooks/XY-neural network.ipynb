```python
import numpy as np
import numpy.linalg as la
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
def derivative(x):
    return x * (1 * x)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
datadir = '../data/processed/'
train_file = datadir + 'train.csv'
train = pd.read_csv(train_file)
```

```python
x = train.values[:, :-1]
y = train.values[:, -1]
```

```python
scaler = StandardScaler().fit(x)
xs = scaler.transform(x)
```

```python
def make_network(n_inputs, n_hidden, neurons_per_hidden):
    input_weights = np.random.random((n_inputs, neurons_per_hidden))
    hidden_weights = [np.random.random((neurons_per_hidden, neurons_per_hidden)) \
                      for _ in range(n_hidden)]
    output_weights = np.random.random((neurons_per_hidden, 1))
    return {
        'input': input_weights,
        'hiddens': hidden_weights,
        'output': output_weights
    }
```

```python
net = make_network(xs.shape[1], 5, 5)
```

```python
net
```

```python
def predict(net, inputs):
    tmp = sigmoid(inputs @ net['input'])
    for h in net['hiddens']:
        tmp = sigmoid(tmp @ h)
    out = sigmoid(tmp @ net['output'])
    return out
```
