```python
from nn4 import NN
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

```python
train = pd.read_csv('data/processed/train.csv')
val = pd.read_csv('data/processed/val.csv')
```

```python
train['NotSpam'] = 1 - train['spam']
val['NotSpam'] = 1 - train['spam']
```

```python
xcols = train.columns[:-2]
ycols = train.columns[-2:]
```

```python
x = train[xcols].values
y = train[ycols].values
s = MinMaxScaler().fit(x)
```

```python
xtr = s.transform(x)
ytr = y
```

```python
x = val[xcols].values
y = val[ycols].values
xv = s.transform(x)
yv = y
```

```python
net = NN(xtr.shape[1], 1, 5, ytr.shape[1])
```

```python
for epoch in range(1):
    for ix in range(xtr.shape[0]):
        net.train(xtr[ix], ytr[ix, 0], lr=1.0)
```
